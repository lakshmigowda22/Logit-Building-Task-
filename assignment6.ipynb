```python
# Problem 1: Smart Parking Lot Management System

def parking_system(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 10

    print("Estimated Delivery Time:", time, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def seat_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

total_seats = 200
booked_seats = [1] * 150  # 150 bookings

seat_analyzer(total_seats, booked_seats)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):

    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def manage_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
manage_registrations(100, 105)
```

```python
def monitor_resources(resource_usage):
    overused = []
    limit = 8

    for resource, hours in resource_usage.items():
        if hours > limit:
            overused.append(resource)

    if overused:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", energy_alert)

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```
