```python
correct_username = "admin"
correct_password = "1234"

username = input("Enter username: ")
password = input("Enter password: ")

if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)

print("\nFinal Result:")
for i in range(len(marks)):
    if marks[i] >= 50:
        print("Student", i+1, "Marks:", marks[i], "-> PASS")
    else:
        print("Student", i+1, "Marks:", marks[i], "-> FAIL")
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    clean = name.strip().lower()
    cleaned_names.append(clean)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "Ok"]

for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    if length > 10:
        print("Status: Longer than 10 characters")
    else:
        print("Status: Short message")

    print("---------------------")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("System Logs:", logs)
print("Total ERROR entries:", error_count)
```
